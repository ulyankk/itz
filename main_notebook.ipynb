```python
import pandas as pd

PATH_DATA = "../data/"
```

```python
df = pd.read_excel(
    PATH_DATA + "Нагрузка.xlsx",
    sheet_name='общее',
)

print(f"Исходное количество строк: {len(df)}")
```

```python
exclude_patterns = [
    'диплом', 'руководство', 'магистр', 'аспирант',
    'преддипломная', 'практика', 'ига', 'гос. экзамен',
    'идпо', 'комисс', 'секрет', 'председ', 'нормоконтрол',
    'резерв', 'отчисл'
]
```

```python
def is_numeric_or_empty(value):
    if pd.isna(value):
        return True
    try:
        pd.to_numeric(value)
        return True
    except (ValueError, TypeError):
        return False
```

```python
mask = (
    df['ФИО'].notna() &  # есть ФИО
    ~df['ФИО'].str.lower().str.contains('э-|ю-|комисс|резерв|пустое|отчисл', na=False, regex=True) &  # исключаем служебные ФИО
    ~df['Вид нагрузки'].str.lower().str.contains('|'.join(exclude_patterns), na=False, regex=True) &  # исключаем виды нагрузки
    df['Нагрузка'].apply(is_numeric_or_empty)  # Нагрузка - число или пусто
)
```

```python
df_regular_load = df[mask].copy()

print(f"Количество строк с обычной нагрузкой: {len(df_regular_load)}")
```

```python
output_file = 'новый_сводный_файл_2026.xlsx'
df_regular_load.to_excel(PATH_DATA + output_file, index=False)
print(f"Результат сохранен в: {output_file}")
```
